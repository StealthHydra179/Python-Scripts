```python
import pandas as pd

path = "./vantagestone-2024-08-27/"
df = pd.read_csv(path + 'vantagestone-2024-08-27.csv',dtype='string', encoding="utf-8")
df1 = df.dropna(subset=["name"],inplace=False)
content_file = open(path + "insert_content.sql", "w",encoding="utf-8")
attribute_file = open(path + "insert_attribute.sql", "w",encoding="utf-8")

#print(df1.head())
#print(df1.info()) 
#df
for i in df1.index:
  #if df1.loc[i, "name"] =="Amani Grey":
    
    #print (df1.loc[i,"name"])
    content_key =  df1.loc[i,"name"].replace (" ","-")
    #print (df1.loc[i,"name"])

    #print(df1.loc[i,"spec"])
    df1.loc[i,"spec"] = "<ul><li>" + df1.loc[i,"spec"].replace ("||","</li><li>") + "</li></ul>"
    #print (df1.loc[i,"spec"])

    #print(df1.loc[i,"description"])
    df1.loc[i,"description"] = "<p>" + df1.loc[i,"description"].replace ("||","</p><p>") + "</p>"
    #print (df1.loc[i,"description"])

    insert_content_part1= "INSERT INTO CONTENT (CONTENT_TYPE_ID, ENABLE_FG, USER_ID, SELLER_ID, PLATFORM_ID, EXPIRY_DATE_TIME, CREATED_DATE_TIME, EFFECTIVE_DATE_TIME, CONTENT_KEY) VALUES (4, 'Y', (SELECT USER_ID FROM USERS WHERE PRIMARY_EMAIL = 'zhangweiheng1213@gmail.com' AND BRANDING_ID = (SELECT BRANDING_ID FROM BRANDING WHERE domain='vantagestone.ca')), (SELECT SELLER_ID FROM SELLER WHERE SELLER_NAME='Vantage Stone Inc.'), (SELECT PLATFORM_ID FROM PLATFORM WHERE ROOT_DOMAIN='vantagestone.ca'), null, now(), now(), '"
    content= insert_content_part1 + df1.loc[i,"name"].replace (" ","-") + "'); \n"
    if not pd.isnull(content):
      print (content)
      content_file.write (content)

    insert_part1= "INSERT INTO CONTENT_ATTRIBUTE (CONTENT_ID, ATTR_KEY, ATTR_SUB_KEY, VALUE, ATTR_SUB_KEY_2) VALUES ((SELECT CONTENT_ID FROM CONTENT WHERE CONTENT_KEY = '" + content_key + "' AND PLATFORM_ID=(SELECT PLATFORM_ID FROM PLATFORM WHERE ROOT_DOMAIN='vantagestone.ca')), "
    attribute0_part2= "'TITLE', '1', '" + df1.loc[i,"name"] + "',null);"
    attribute0 = insert_part1 + attribute0_part2 + " \n"
    if not pd.isnull(attribute0):
      print (attribute0)
      attribute_file.write (attribute0)

    attribute1_part2= "'SECTION1', '1', '" + df1.loc[i,"spec"] + "',null);"
    attribute1 = insert_part1 + attribute1_part2 + " \n"
    if not pd.isnull(attribute1):
      print (attribute1)
      attribute_file.write (attribute1)

    attribute2_part2= "'SECTION2', '1', '" + df1.loc[i,"description"] + "',null);"
    attribute2 = insert_part1 + attribute2_part2 + " \n"
    if not pd.isnull(attribute2):
      print (attribute2)
      attribute_file.write (attribute2)

    # similar atrributes continue...
    
content_file.close()   
attribute_file.close()   
```
